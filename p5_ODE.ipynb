# ODE

## Ejercicio 1
$y'= -5y+5t^2+2t$
* $0 \le t \le 1$
* $y(t_0)=1/3$
* $N=10$



```python
%autoreload 2
from src import ODE_euler
```

## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



```python
import matplotlib.pyplot as plt
import numpy as np
from src import ODE_euler, ODE_euler_nth

f = lambda t, y: -5*y + 5*t**2 + 2*t
y_t0 = 1/3
a = 0
b = 1
ys,ts, h = ODE_euler(a=a, b=b, y_t0=y_t0, f=f, N=10)
ys

y_real = lambda t: t**2 + 1 / 3 * np.exp(-5 * t)

print("El valor de h es: ", h)

#Grafica   
# real
t_real = np.linspace(a, b, 100)
y_real = [y_real(t) for t in t_real]
plt.plot(t_real, y_real, label="Solucion real")

# Aproximada
plt.plot(ts, ys, marker=".", linestyle=":", label=r"Solucion aproximada")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Comparacion con la solucion real")
plt.legend()
plt.grid()
plt.show()
```

# Errores
Calcule el eeror relativo para diferentes valores de N.

### N = 10

```python

# Calcular el error relativo
error_relativo = [abs(y_real_val - y_approx) / abs(y_real_val) 
                  for y_real_val, y_approx in zip(y_real, ys)]

print(error_relativo)

error_relativo_promedio = sum(error_relativo) / len(error_relativo)
print(f"Error relativo promedio: {error_relativo_promedio}")
```
